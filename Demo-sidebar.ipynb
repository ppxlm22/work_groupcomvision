# การสร้าง Slide Bar

```python
import cv2
import numpy as np

cap = cv2.VideoCapture(0)
cv2.namedWindow("HSV Turner")
cv2.createTrackbar("H Low", "HSV Turner", 0, 179, lambda x: None)
cv2.createTrackbar("H High", "HSV Turner", 179, 179, lambda x: None)
cv2.createTrackbar("S low","HSV Turner",0,255 ,lambda x: None)
cv2.createTrackbar("S high","HSV Turner",255,255,lambda x: None)
cv2.createTrackbar("V low","HSV Turner",0,255 ,lambda x: None)
cv2.createTrackbar("V high","HSV Turner",255,255,lambda x: None)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    H_Low = cv2.getTrackbarPos("H Low","HSV Turner")
    H_High = cv2.getTrackbarPos("H High" , "HSV Turner")
    S_Low = cv2.getTrackbarPos("S low","HSV Turner")
    S_High = cv2.getTrackbarPos("S high","HSV Turner")
    V_Low = cv2.getTrackbarPos("V low","HSV Turner")
    V_High = cv2.getTrackbarPos("V high","HSV Turner")

    HSV = cv2.cvtColor(frame,cv2.COLOR_BGR2HSV)
    lower = (H_Low , S_Low , V_Low)
    upper = (H_High, S_High , V_High)
    mask = cv2.inRange(HSV,lower,upper)
    
    result = cv2.bitwise_and(frame,frame,mask=mask)
    result = cv2.erode(result, None, iterations=2)
    result = cv2.dilate(result, None, iterations=2)

    # แปลง mask เป็น 3 channels เพื่อรวมภาพ
    mask_3ch = cv2.cvtColor(mask, cv2.COLOR_GRAY2BGR)
    
    # รวมภาพแบบ 2x2
    top_row = np.hstack((frame, HSV))
    bottom_row = np.hstack((mask_3ch, result))
    combined = np.vstack((top_row, bottom_row))
    
    cv2.imshow("HSV Turner", combined)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
